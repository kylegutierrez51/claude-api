````python
import os
from dotenv import load_dotenv

load_dotenv()
````

````python
from anthropic import Anthropic

client = Anthropic()
model = os.getenv("CLAUDE_MODEL")
````

````python
def add_user_message(messages, text):
	user_message = {"role": "user", "content": text}
	messages.append(user_message)


def add_assistant_message(messages, text):
	assistant_message = {"role": "assistant", "content": text}
	messages.append(assistant_message)


def chat(messages, system=None, temperature=1.0, stop_sequences=None):
	params = {
		"model" : model,
		"max_tokens" : 300,
		"messages" : messages,
		"temperature" : temperature,
		"stop_sequences" : stop_sequences,
	}

	if system:
		params["system"] = system
	
	message = client.messages.create(**params)

	return message.content[0].text
````

Exercise:
 - Use message prefilling and stop sequences only to get three different commands in a single response
 - There shouldn't be any comments or explanation
 - Hint: message prefilling isn't limited to just characters like ```

````python
# Starting code

messages = []

prompt = """
	Generate three different sample AWS CLI commands. Each should be very short.
"""

add_user_message(messages, prompt)

text = chat(messages)
text.strip()
````

````python
# Finished code
# Inherently flawed solution, infact, stop sequences as a whole seem completely flawed since Claude may not generate those backticks "```"
# Sometimes when you run you get just one command outputted. Not reliable. It's probably better to take the text and perform an algorithm that can just send back the 3 commands by themselves.
messages = []

prompt = """
	Generate three different sample AWS CLI commands. Each should be very short.
"""

add_user_message(messages, prompt)
add_assistant_message(messages, "Here are all three commands in a single block without any comments:\n```bash")

text = chat(messages, stop_sequences=["```"])
text.strip()
````
